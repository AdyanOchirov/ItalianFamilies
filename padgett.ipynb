```python
import numpy as np
import xml.etree.ElementTree as ET
tree = ET.parse('padgett.xml')
root = tree.getroot()
```

```python
i=0
nodes=np.empty(16, dtype=object)
for child in root[0][0][0]:
    nodes[i]=child.get('id')
    i=i+1
```

```python
i=0
links=np.empty(16*16, dtype=object)
for link in root[0][1][0].iter('link'):
    links[i]=int(float(link.get('value')))
    i=i+1
links=links.reshape(16,16)
```

```python
p=(links.T/(links.sum(axis=0)+(links.sum(axis=0)==0))).T
```

```python
ranksnext=np.ones(16)/16
ranks=np.ones(16)
while ((np.abs(ranksnext-ranks)>np.full(16,0.0001)).any()):
    ranks=ranksnext
    ranksnext=np.matmul(ranksnext,p)
```

```python
ans=np.concatenate((nodes[:,None],ranksnext[:,None]),axis=1)
ans[(-ans[:, 1]).argsort()]
```
